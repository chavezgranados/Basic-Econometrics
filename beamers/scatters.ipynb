#**Diagramas de dispersi√≥n**

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(2000)  # Para reproducibilidad
n_points = 100

# 1. Lineal positivo
x1 = np.linspace(0, 10, n_points)
y1 = 2 * x1 + 1 + np.random.normal(0, 3, n_points)

# 2. Lineal negativo
x2 = np.linspace(0, 10, n_points)
y2 = -1.5 * x2 + 8 + np.random.normal(0, 3, n_points)

# 3. Cuadr√°tico
x3 = np.linspace(-5, 5, n_points)
y3 = x3**2 + np.random.normal(0, 5, n_points)

# 4. C√∫bico
x4 = np.linspace(-3, 3, n_points)
y4 = x4**3 - 2 * x4 + np.random.normal(0, 10, n_points)

# 5. Exponencial
x5 = np.linspace(0, 5, n_points)
y5 = np.exp(x5) + np.random.normal(0, 5, n_points)

# 6. Logar√≠tmico
x6 = np.linspace(1, 10, n_points)
y6 = np.log(x6) + np.random.normal(0, 0.4, n_points)

# 7. Rec√≠proco
x7 = np.linspace(1, 10, n_points)
y7 = 1 / x7 + np.random.normal(0, 0.2, n_points)

# 8. Otro (ejemplo: sinusoidal)
x8 = np.linspace(0, 10, n_points)
y8 = np.sin(x8) + np.random.normal(0, 0.5, n_points)

# Crear la figura y los subplots
fig, axs = plt.subplots(4, 2, figsize=(12, 16))
axs = axs.flatten()

# Datos y colores
data = [(x1, y1, 'red'), (x2, y2, 'blue'), (x3, y3, 'green'), (x4, y4, 'orange'),
         (x5, y5, 'purple'), (x6, y6, 'brown'), (x7, y7, 'cyan'), (x8, y8, 'magenta')]

# Graficar cada diagrama de dispersi√≥n
for i, (x, y, color) in enumerate(data):
  axs[i].scatter(x, y, color=color, alpha=0.7)

plt.tight_layout()
plt.show()
```
